```python
class Categorical_Extractor(BaseEstimator, TransformerMixin):
    """Get dummy variables for categorical features"""
    
    def __init__(self):
        pass

    def get_dummy_(self,df):
        
        icu_dummy = pd.get_dummies(df['icutype'],drop_first=True)
        df = pd.concat((df['gender'],icu_dummy),axis=1)
        
        return df.values.reshape(-1, 4)
    
    def transform(self, X, *args):
        
        X = self.get_dummy_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
class Numerical_Extractor(BaseEstimator, TransformerMixin):
    """Use RobustScaler on numerical features"""
    
    def __init__(self):
        pass

    def robust_scale_(self,df,threshold=10):
        
        num_col = []
        for col, num in df.nunique().iteritems():
            if num > threshold:
                num_col.append(col)
                
        df = df[num_col]
        
        rscaler = RobustScaler()
        array = rscaler.fit_transform(df)
        
        return array.reshape(-1,array.shape[1])
    
    def transform(self, X, *args):
        
        X = self.robust_scale_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
feature_union = FeatureUnion([  
    ('numerical_scaler',Numerical_Extractor()),
    ('categorical_transform',Categorical_Extractor())
])
```

```python
transformed_X = feature_union.transform(X)
```

```python
class Clean_FeatureUnion(BaseEstimator, TransformerMixin):
    """Convert array from feature union into dataframe"""
    
    def __init__(self):
        pass
    
    
    def array_to_df(self,array):
        
        new_columns = list(X.columns)
        new_columns.extend(['icutype3','icutype4'])
        
        X_df = pd.DataFrame(array,columns=new_columns,index=X.index)
        X_df = X_df.rename(columns={'icutype':'icutype2'})
        
        return X_df
    
    def transform(self, union_array, *args):
        
        df = self.array_to_df(union_array)
        
        return df


    def fit(self, union_array, *args):
        return self
```

<b> Variance Inflation Factor</b>

</div>

```python
# Variance Inflation Factor

class VIF(BaseEstimator, TransformerMixin):
    """Variance Inflation Factor"""
    
    def __init__(self):
        pass
    
    
    def vif(self,df):
        
        vif = [variance_inflation_factor(df.iloc[:,:-4].values,i) \
               for i in range(df.iloc[:,:-4].shape[1])]
        
        vif_df = pd.DataFrame(df.iloc[:,:-4].columns,columns=['Features'])
        vif_df['VIF'] = vif    # VIF values in dataframe
        
        remove_col = list(vif_df[vif_df['VIF']>5]['Features'])   # Choose only features with VIF < 5
        selected_df = df.drop(remove_col,axis=1)
        
        return selected_df
    
    def transform(self, df, *args):
        
        selected_df = self.vif(df)
        
        return selected_df


    def fit(self, df, *args):
        return self
```
